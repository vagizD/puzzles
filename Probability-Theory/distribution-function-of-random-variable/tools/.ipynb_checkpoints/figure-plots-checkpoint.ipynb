### Plots for the puzzle solution

```python
import matplotlib.pyplot as plt
import numpy as np

%config InlineBackend.figure_format = 'retina'
```

```python
fig, ax = plt.subplots(figsize=(6,6))


ax.set(xlim=(0,1),
       ylim=(0,1),
       title="Empty grid")

ax.grid(b=True, alpha=0.7, ls="--")

ax.scatter(x=[0.55], y=[0.65], s=20)
ax.text(x=0.575, y=0.675, s="(w1,w2)")

fig.savefig('empty-grid.png', bbox_inches='tight')
```

```python
fig, ax = plt.subplots(figsize=(6,6))


ax.set(xlim=(0,1),
       ylim=(0,1),
       title="Obvious notice",
       xlabel="w1",
       ylabel="w2")

ax.grid(b=True, alpha=0.7, ls="--")


xx = np.linspace(0,1,100)
yy = [1-x for x in xx]

ax.plot(xx, yy, ls='--')
ax.text(0.2, 0.7, s="w1 + w2 <= 1", rotation=-45)
ax.fill_between(x=xx, y1=yy, y2=0, color='yellow')


ax.scatter([1/2], [1/2], s=20, color='red')
ax.text(1/2 + 0.025, 1/2 + 0.025, s="(1/2, 1/2)")


fig.savefig('obvious-notice.png', bbox_inches='tight', dpi='figure')
```

```python
fig, ax = plt.subplots(figsize=(6,6))


ax.set(xlim=(0,1),
       ylim=(0,1),
       title="x in [0,1]",
       xlabel="w1",
       ylabel="w2")

ax.grid(b=True, alpha=0.7, ls="--")


xx = np.linspace(0, 0.5, 100)
yy = [0.5 - x for x in xx]

ax.plot(xx, yy, ls='--')
ax.text(0.05 + 0.025, 0.05 + 0.025, s="w1 + w2 <= x for x in [0,1]", rotation=-45)
ax.fill_between(x=xx, y1=yy, y2=0, color='yellow')

ax.scatter(x=[0.25], y=[0.25], s=20, color='red')
ax.text(0.25 + 0.025, 0.25 + 0.025, s="(x/2,x/2)")

fig.savefig('below-line.png', bbox_inches='tight', dpi='figure')
```

```python
fig, ax = plt.subplots(figsize=(6,6))


ax.set(xlim=(0,1),
       ylim=(0,1),
       title="x in [1,2]",
       xlabel="w1",
       ylabel="w2")

ax.grid(b=True, alpha=0.7, ls="--")


xx = np.linspace(0.5, 1, 100)
yy = [1.5-x for x in xx]

ax.plot(xx, yy, ls='--')
ax.text(0.55 + 0.025, 0.55 + 0.025, s="w1 + w2 <= x for x in [1,2]", rotation=-45)
ax.fill_between(x=xx, y1=yy, y2=0, color='yellow')
ax.fill_between(x=[0,0.5], y1=1,y2=0, color='yellow')

ax.scatter(x=[0.75], y=[0.75], s=20, color='red')
ax.text(0.75 + 0.025, 0.75 + 0.025, s="(x/2,x/2)")

fig.savefig('above-line.png', bbox_inches='tight', dpi='figure')
```
