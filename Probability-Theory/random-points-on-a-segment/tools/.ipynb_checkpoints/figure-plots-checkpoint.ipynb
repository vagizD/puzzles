### Plots for the puzzle solution

```python
import matplotlib.pyplot as plt
import numpy as np

%config InlineBackend.figure_format = 'retina'

fig, ax = plt.subplots(figsize=(6,4))

ax.spines['bottom'].set_position(('center'))
ax.spines['left'].set_position(('center'))
ax.spines[['right', 'top']].set_visible(False)

plt.xticks([])
plt.yticks([]);

ax.set(xlim=(-10, 10),
       ylim=(-10,10))

ax.axhline(y=10, xmin=0, xmax=1)
ax.axhline(y=-10, xmin=0, xmax=1)
ax.axvline(x=10, ymin=0, ymax=1)
ax.axvline(x=-10, ymin=0, ymax=1)


ax.text(-11.5, -0.5, "-n", fontsize='xx-large')
ax.text(-0.5, -11.5, "-n", fontsize='xx-large')
ax.text(10.5, -0.5, "n", fontsize='xx-large')
ax.text(-0.5, 10.5, "n", fontsize='xx-large')

ax.scatter([5], [5], c='blue', s=20)
ax.text(5, 6, "(p,q)")



fig.suptitle("Segment --> Plane")
fig.savefig('segment-plane.png', bbox_inches='tight')
```

```python
fig, ax = plt.subplots(figsize=(6,4))

ax.spines['bottom'].set_position(('center'))
ax.spines['left'].set_position(('center'))
ax.spines[['right', 'top']].set_visible(False)

plt.xticks([])
plt.yticks([]);

ax.set(xlim=(-10, 10),
       ylim=(-10,10))

ax.axhline(y=10, xmin=0, xmax=1)
ax.axhline(y=-10, xmin=0, xmax=1)
ax.axvline(x=10, ymin=0, ymax=1)
ax.axvline(x=-10, ymin=0, ymax=1)


ax.text(-11.5, -0.5, "-n", fontsize='xx-large')
ax.text(-0.5, -11.5, "-n", fontsize='xx-large')
ax.text(10.5, -0.5, "n", fontsize='xx-large')
ax.text(-0.5, 10.5, "n", fontsize='xx-large')


xx = np.linspace(-2*np.sqrt(10), 2*np.sqrt(10), 1000)
yy = [(p**2)/4 for p in xx]
plt.plot(xx, yy)

ax.fill_between(xx, yy, 0, color='orange')


xx = np.linspace(-10, 10, 1000)
yy = np.linspace(-10, -10, 1000)
ax.fill_between(xx, yy, color='red')

xx = np.linspace(-10, -2*np.sqrt(10), 100)
yy = [(p**2)/4 for p in xx]
ax.fill_between(xx, yy, 0, color='green')
xx = np.linspace(2*np.sqrt(10), 10, 100)
yy = [(p**2)/4 for p in xx]
ax.fill_between(xx, yy, 0,  color='green')

fig.suptitle("Square break down")
fig.savefig('square-break-down.png', bbox_inches='tight')
```
